## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# --- 11A: Drop columns that are identifiers or near-completely missing ---
cols_to_drop = ['Unique id', 'connected_handling_time', 'Order_id', 'order_date_time', 'Survey_response_Date']
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
print(f"\n Dropped columns: {cols_to_drop}")
````

````python
# --- 11B: Parse datetime columns ---
df['Issue_reported at'] = pd.to_datetime(df['Issue_reported at'], dayfirst=True, errors='coerce')
df['issue_responded']   = pd.to_datetime(df['issue_responded'],   dayfirst=True, errors='coerce')
print("Datetime columns parsed.")
````

````python
# --- 11C: Create Response Time feature (minutes) ---
df['response_time_mins'] = (
    df['issue_responded'] - df['Issue_reported at']
).dt.total_seconds() / 60
# Cap extreme negatives/positives
df['response_time_mins'] = df['response_time_mins'].clip(lower=0, upper=300)
print(" Response time (minutes) feature created.")
````

````python
df['reported_hour']     = df['Issue_reported at'].dt.hour
df['reported_dayofweek']= df['Issue_reported at'].dt.day_name()
df['reported_month']    = df['Issue_reported at'].dt.month_name()
print(" Time-based features extracted (hour, day, month).")
````

````python
# --- 11E: Handle optional columns with high missingness ---
# Fill with 'Unknown' — keeps rows, flags the missingness
for col in ['Customer_City', 'Product_category', 'Customer Remarks']:
    if col in df.columns:
        df[col].fillna('Unknown', inplace=True)

# Item_price: fill with median
if 'Item_price' in df.columns:
    median_price = df['Item_price'].median()
    df['Item_price'].fillna(median_price, inplace=True)
    print(f"Item_price nulls filled with median: {median_price:.2f}")
````

````python
# --- 11F: Strip whitespace from string columns ---
str_cols = df.select_dtypes(include='object').columns
for col in str_cols:
    df[col] = df[col].str.strip()
print(" Whitespace stripped from string columns.")
````

````python
# --- 11G: Standardize CSAT Score (keep only valid 1–5) ---
df = df[df['CSAT Score'].between(1, 5)]
print(f" CSAT Score filtered to 1–5. Rows remaining: {len(df)}")
````

````python
# --- 11H: Final check ---
print(f"\nData Wrangling Complete. Final shape: {df.shape}")
print(f"   Remaining nulls:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
````

### What all manipulations have you done and insights you found?

Answer:
1.	Dropped high-null columns: connected_handling_time (99% nulls), Order_id (identifier), Unique id (identifier), Survey_response_Date (not analytically useful).
2.	Datetime parsing: Issue_reported at and issue_responded were converted to datetime format.
3.	Response time feature created: response_time_mins = (issue_responded − Issue_reported_at).dt.total_seconds() / 60 — a critical derived feature.
4.	Time-based features extracted: issue_hour, issue_day_of_week, issue_month from Issue_reported at.
5.	Categorical nulls filled with mode; numerical nulls filled with median.
6.	Label Encoding applied to categorical columns for correlation analysis.
7.	Binary satisfaction flag created: satisfied = 1 if CSAT ≥ 4, else 0.
8.	CSAT Score filtered to valid range 1–5 to remove any data entry errors.
Key insight from wrangling: The order_date_time column had 80% nulls — filling with mode or extracting features from it introduces significant noise. Issue_reported at is a more reliable datetime column and was used for all time-based engineering.
